```python
#Created 2 seprate objects of petal length and petal width
pl = iris.PetalLengthCm  #pl is petal length
pw = iris.PetalWidthCm   #pw is petal width
```

```python
#Plotted a graph of petal length and petal width.
#plt.style.use('ggplot')
plt.plot(pl,'b*',label='Petal Length')
plt.plot(pw,'g*', label='Petal width')
plt.title('Iris Data-Set')
plt.xlabel('No of Petals')
plt.ylabel('Dimension in cm')
plt.legend()
plt.show()
```

```python
#Created 2 seprate objects of sepal length and sepal width
sl = iris.SepalLengthCm  #sl is sepal length 
sw = iris.SepalWidthCm  #sw is sepal width 
```

```python
#Plotted a graph of sepal length and sepal width.
#plt.style.use('ggplot')
plt.plot(sl,'b*',label='Sepal Length')
plt.plot(sw,'g*', label='Sepal width')
plt.title('Iris Data-Set')
plt.xlabel('No of Sepal')
plt.ylabel('Dimension in cm')
plt.legend()
plt.show()
```

```python
# Created three objects of each species.
#Seperated these three species with their opject i.e ir_set,ir_ver,ir_vir
ir_ver = iris[iris.Species == 'Iris-versicolor']
ir_set = iris[iris.Species == 'Iris-setosa']
ir_vir = iris[iris.Species == 'Iris-virginica']
```

```python
# Decriebe() Function gives the mean, std ,max etc.
print(ir_ver.describe())
print(ir_set.describe())
print(ir_vir.describe())
```

```python
#Now, created object of each species with their petal length and width as well as sepal length and width.
# These objects are of Iris-versicolor
ver_pl = ir_ver.PetalLengthCm #Iris-versicolor petal length
ver_pw = ir_ver.PetalWidthCm  #Iris-versicolor petal width
ver_sl = ir_ver.SepalLengthCm #Iris-versicolor sepal length
ver_sw = ir_ver.SepalWidthCm  #Iris-versicolor sepal width
```

```python
# Plotted a scatter plot of iris-versicolor blue dots are sepal and red are petal.
plt.scatter(ver_pl,ver_pw,label="Petal")
plt.scatter(ver_sl,ver_sw,label="Sepal")
plt.title('Iris-versicolor')
plt.xlabel('length')
plt.ylabel('width')
plt.legend()
plt.show()
```

```python
# These objects are of Iris-setosa
set_pl = ir_set.PetalLengthCm # Iris-setosa petal length
set_pw = ir_set.PetalWidthCm  #Iris-setosa petal width
set_sl = ir_set.SepalLengthCm #Iris-setosa sepal length
set_sw = ir_set.SepalWidthCm  #Iris-setosa sepal width
```

```python
# Plotted a scatter plot of iris-setosa blue dots are sepal and red are petal.
plt.scatter(set_pl,set_pw,label="Petal")
plt.scatter(set_sl,set_sw,label="Sepal")
plt.title('Iris-setosa')
plt.xlabel('length')
plt.ylabel('width')
plt.legend()
plt.show()
```

```python
# These objects are of Iris-virginica
vir_pl = ir_ver.PetalLengthCm #Iris-virginica petal length
vir_pw = ir_vir.PetalWidthCm  #Iris-virginica petal width
vir_sl = ir_vir.SepalLengthCm #Iris-virginica sepal length
vir_sw = ir_vir.SepalWidthCm  #Iris-virginica sepal width
```

```python
# Plotted a scatter plot of iris-virginica blue dots are sepal and red are petal.
plt.scatter(vir_pl,vir_pw,label="Petal")
plt.scatter(vir_sl,vir_sw,label="Sepal")
plt.title('Iris-virginica')
plt.xlabel('length')
plt.ylabel('width')
plt.legend()
plt.show()
```

```python
# this scatter plot of all species with their petal length-----------

plt.scatter(ver_pl,ver_pw,color='#443283', label="Versicolor")
plt.scatter(set_pl,set_pw,color='#f73498',label="Setosa")
plt.scatter(vir_pl,vir_pw,color='#f99534',label="Virginica")
plt.legend()
plt.title("Petal length Scatter Plot of -\nSetosa, Versicolour, and Virginica")
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

```python
# this scatter plot of all species with their sepal length-----------
plt.scatter(ver_sl,ver_sw,color='c',label="Versicolor")
plt.scatter(set_sl,set_sw,color='r',label="Setosa")
plt.scatter(vir_sl,vir_sw,color='y',label="Virginica")
plt.legend()
plt.title("Sepal length Scatter Plot of -\nSetosa, Versicolour, and Virginica")
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```
